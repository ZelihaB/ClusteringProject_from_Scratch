### RFM

```python
rfm = df.groupby('Customer ID').agg({'Recency': lambda date: (today_date - date.max()).days,
                                     'Invoice': lambda num: num.nunique(),
                                     'Revenue': lambda Revenue: Revenue.sum().astype(int),
                                    'Tenure': lambda date: (today_date - date.min()).days})
```

```python
rfm.head()
```

```python
rfm.columns
```

```python
#Rename columns same as literature
rfm.columns = ['Recency', 'Frequency', 'Monetary','Tenure']

rfm.head()
```

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])  #columnname, how many splits and the labels
```

```python
rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm["TenureScore"] = pd.qcut(rfm['Tenure'], 5, labels=[1, 2, 3, 4, 5])
```

```python
pd.crosstab(rfm["Frequency"],rfm["FrequencyScore"])
```

```python
rfm.head()
```

```python
###############################################################
# Naming & Analysing RFM Segments
###############################################################

# RFM Labeling
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At_Risk',
    r'[1-2]5': 'Cant_Loose',
    r'3[1-2]': 'About_to_Sleep',
    r'33': 'Need_Attention',
    r'[3-4][4-5]': 'Loyal_Customers',
    r'41': 'Promising',
    r'51': 'New_Customers',
    r'[4-5][2-3]': 'Potential_Loyalists',
    r'5[4-5]': 'Champions'
}

```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm.head()
```

```python
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
df[["Customer ID"]].nunique()
```
